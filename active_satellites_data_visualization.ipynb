```python
import matplotlib.pyplot as plt
```

```python
brazil_satellites = df[df['country/org_UNreg'] == 'BRA']
brazil_total_satellites = len(brazil_satellites)
print(f'Brazilian satellites: {brazil_total_satellites}')
```

```python
purpose_counts = brazil_satellites['purpose'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=purpose_counts.index, y=purpose_counts.values, palette='viridis')
plt.title('Purpose of Brazilian Satellites')
plt.ylabel('Number of Satellites')
plt.xlabel('Purpose')
plt.xticks(ha='right')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['purpose'], bins=50, kde=True)
plt.title('Distribution of Purpose')
plt.xlabel('Purpose')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['purpose'])
plt.title("Satellites purpose distribution")
plt.xlabel("Purpose")
plt.xticks(rotation=90)
plt.show()
```

```python
import plotly.express as px
usa_satellites = df[df['country/org_UNreg'] == 'USA']
usa_total_satellites = len(usa_satellites)
print(f'USA total satellites: {usa_total_satellites}')

fig = px.pie(names=usa_satellites['purpose'].value_counts().index,
             values=usa_satellites['purpose'].value_counts().values,
             title="USA satellites purpose proportion",
             hole=0.25)
fig.update_traces(textinfo='percent+label')
fig.show()
```
